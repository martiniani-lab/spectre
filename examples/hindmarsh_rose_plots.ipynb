```python
"""We define the class for simulating the Hindmarsh-Rose model."""

import torch
import matplotlib.pyplot as plt
from spectre.spectrum_general import matrix_solution
from spectre.spectrum_general import sim_solution
from spectre.spectrum_general import element_wise
from spectre.spectrum_general import recursive_solution
from spectre.model import HR
import numpy as np


plt.rc("text", usetex=True)
plt.rc("font", family="serif")
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device = torch.device("cpu")

fileloc = r"../figures/HR/"
```

```python
"""Define the colors for the different methods"""

color_recursive = "#FFA353"
color_elementwise = "#095400"
color_matrix = "#1B3B9D"
color_sim = "k"

linewidth_recursive = 12
linewidth_elementwise = 3.5
marker_matrix = 8
simulation_marker = 12

font_size = 28
```

```python
"""Define the HR circuit"""

eta1 = 0.001
I = 5.5
model = HR(eta1=eta1, I=I)
```

```python
"""Define the simulation parameters"""

ndivs = 10
n_points = int(1e6)
time = 50000
dt = 0.04
```

```python
"""Define the frequency vectors"""

min_freq = 1e-4
max_freq = 10
n_freq_mat = 60
n_freq_rat = 1000

"""Define the frequency vectors for the matrix and rational solutions"""
freq_mat = torch.logspace(np.log10(min_freq), np.log10(max_freq), n_freq_mat)
freq_rat = torch.logspace(np.log10(min_freq), np.log10(max_freq), n_freq_rat)
```

```python
"""Find auto-spectra of x using different schemes"""

idx = 0

"""Analytical solution (matrix)"""
mat_sol = matrix_solution(model.J, model.L, model.S)
psd_matrix, _ = mat_sol.auto_spectrum(idx, freq_mat)

"""Analytical solution (elementwise)"""
rat_sol = element_wise(model.J, model.L, model.S)
psd_rational, _ = rat_sol.auto_spectrum(idx, freq_rat)

"""Analytical solution (recursive algorithm)"""
recursive_sol = recursive_solution(model.J, model.L, model.S)
psd_recursive, _ = recursive_sol.auto_spectrum(idx, freq_rat)

"""Simulation solution"""
sim_sol = sim_solution(model)
psd_sim, f = sim_sol.simulation_spectrum(
    i=idx, ndivs=ndivs, n_points=n_points, time=time, dt=dt
)
```

```python
plt.figure(1)
idx_closest = torch.unique(torch.abs(f[1:, None] - freq_mat).argmin(axis=0)) + 1
plt.loglog(
    freq_rat,
    psd_recursive,
    label="recursive",
    linewidth=linewidth_recursive,
    color=color_recursive,
)
plt.loglog(
    freq_rat,
    psd_rational,
    label="element-wise",
    linewidth=linewidth_elementwise,
    color=color_elementwise,
)
plt.loglog(
    freq_mat,
    psd_matrix,
    label="matrix",
    linewidth=0,
    marker="s",
    markerfacecolor="none",
    markersize=marker_matrix,
    color=color_matrix,
)
plt.loglog(
    f[idx_closest],
    psd_sim[idx_closest],
    label="simulation",
    linewidth=0,
    marker="o",
    markersize=simulation_marker,
    markerfacecolor="none",
    color=color_sim,
)
plt.legend(fontsize=18, loc="lower left")
plt.xlabel("Frequency (Hz)", fontsize=20)
plt.ylabel("Power Spectral Density", fontsize=20)
plt.xticks([1e-3, 1e-1, 10], ["0.001", "0.1", "10"], fontsize=20)
plt.yticks([1e-11, 1e-7, 1e-3], fontsize=20)
plt.xlim([1e-4, 10])
plt.ylim([1e-11, 1e-3])
plt.tight_layout()
plt.savefig(fileloc + "psd_hr.svg", format="svg", dpi=1200)
plt.show()
```

```python
"""Plot the simulated spectrum of the membrane potential and the one using the formula"""

idx = 0

"""Simulation spectrum"""
# sim_sol = sim_solution(model)
psd_sim, f = sim_sol.simulation_spectrum(
    i=idx, ndivs=ndivs, n_points=n_points, time=time, dt=dt
)

"""Formula spectrum"""
f_psd = model.psd_x(freq_rat)
```

```python
plt.figure(2, figsize=(7, 5))
idx_closest = torch.unique(torch.abs(f[1:, None] - freq_mat).argmin(axis=0)) + 1
plt.loglog(freq_rat, f_psd, label="analytical", linewidth=6, color="C1")
plt.loglog(
    f[idx_closest],
    psd_sim[idx_closest],
    label="simulation",
    linewidth=0,
    marker="o",
    markersize=simulation_marker,
    markerfacecolor="none",
    color=color_sim,
)
plt.legend(fontsize=18, loc="lower left")
plt.xlabel("Frequency (Hz)", fontsize=20)
plt.ylabel("Power Spectral Density", fontsize=20)
plt.xticks([1e-3, 1e-1, 10], ["0.001", "0.1", "10"], fontsize=20)
plt.yticks([1e-11, 1e-7, 1e-3], fontsize=20)
plt.xlim([1e-4, 10])
plt.ylim([1e-11, 1e-3])
plt.tight_layout()
plt.savefig(fileloc + "psd_hr_formula.svg", format="svg", dpi=1200)
plt.show()
```

```python
"""Plot the cross-spectrum between x and y"""

i = 0
j = 1

"""Analytical solution (matrix)"""
# mat_sol = matrix_solution(model.J, model.L, model.S)
cpsd_matrix, _ = mat_sol.cross_spectrum(i, j, freq_mat)

"""Analytical solution (elementwise)"""
# rat_sol = element_wise(model.J, model.L, model.S)
cpsd_rational, _ = rat_sol.cross_spectrum(i, j, freq_rat)

"""Analytical solution (recursive algorithm)"""
# recursive_sol = recursive_solution(model.J, model.L, model.S)
cpsd_recursive, _ = recursive_sol.cross_spectrum(i, j, freq_rat)

"""Simulation solution"""
# sim_sol = sim_solution(model)
cpsd_sim, f = sim_sol.simulation_spectrum(
    i=i, j=j, ndivs=ndivs, n_points=n_points, time=time, dt=dt
)
```

```python
plt.figure(3)
idx_closest = torch.unique(torch.abs(f[1:, None] - freq_mat).argmin(axis=0)) + 1
plt.semilogx(
    freq_rat,
    torch.abs(cpsd_recursive),
    label="recursive",
    linewidth=linewidth_recursive,
    color=color_recursive,
)
plt.semilogx(
    freq_rat,
    torch.abs(cpsd_rational),
    label="element-wise",
    linewidth=linewidth_elementwise,
    color=color_elementwise,
)
plt.semilogx(
    freq_mat,
    torch.abs(cpsd_matrix),
    label="matrix",
    linewidth=0,
    marker="s",
    markerfacecolor="none",
    markersize=marker_matrix,
    color=color_matrix,
)
plt.semilogx(
    f[idx_closest],
    torch.abs(cpsd_sim[idx_closest]),
    label="simulation",
    linewidth=0,
    marker="o",
    markersize=simulation_marker,
    markerfacecolor="none",
    color=color_sim,
)
plt.legend(fontsize=18, loc="upper right")
plt.xlabel("Frequency (Hz)", fontsize=20)
plt.ylabel("Cross Power Spectral Density", fontsize=20)
plt.xticks([1e-3, 1e-1, 10], ["0.001", "0.1", "10"], fontsize=20)
plt.xlim([1e-4, 10])
plt.yticks([1e-6, 3e-6, 5e-6], fontsize=20)
# plt.ylim([1e-10, 1e-5])
plt.tight_layout()
plt.savefig(fileloc + "cpsd_hr_absolute.svg", format="svg", dpi=1200)
plt.show()
```

```python
"""Plot the coherence between x and y"""

i = 0
j = 1

"""Analytical solution (matrix)"""
# mat_sol = matrix_solution(model.J, model.L, model.S)
coh_matrix, _ = mat_sol.coherence(i=i, j=j, freq=freq_mat)

"""Analytical solution (elementwise)"""
# rat_sol = element_wise(model.J, model.L, model.S)
coh_rational, _ = rat_sol.coherence(i, j, freq_rat)

"""Analytical solution (recursive algorithm)"""
# recursive_sol = recursive_solution(model.J, model.L, model.S)
coh_recursive, _ = recursive_sol.coherence(i, j, freq_rat)

"""Simulation solution"""
# sim_sol = sim_solution(model)
coh_sim, f = sim_sol.simulation_coherence(
    i=i, j=j, ndivs=ndivs, n_points=n_points, time=time, dt=dt
)
```

```python
plt.figure(4)
idx_closest = torch.unique(torch.abs(f[1:, None] - freq_mat).argmin(axis=0)) + 1
plt.semilogx(
    freq_rat,
    torch.abs(coh_recursive),
    label="recursive",
    linewidth=linewidth_recursive,
    color=color_recursive,
)
plt.semilogx(
    freq_rat,
    torch.abs(coh_rational),
    label="element-wise",
    linewidth=linewidth_elementwise,
    color=color_elementwise,
)
plt.semilogx(
    freq_mat,
    torch.abs(coh_matrix),
    label="matrix",
    linewidth=0,
    marker="s",
    markerfacecolor="none",
    markersize=marker_matrix,
    color=color_matrix,
)
plt.semilogx(
    f[idx_closest],
    torch.abs(coh_sim[idx_closest]),
    label="simulation",
    linewidth=0,
    marker="o",
    markersize=simulation_marker,
    markerfacecolor="none",
    color=color_sim,
)
plt.legend(fontsize=18, loc="lower right")
plt.xlabel("Frequency (Hz)", fontsize=20)
plt.ylabel("Coherence", fontsize=20)
plt.xticks([1e-3, 1e-1, 10], ["0.001", "0.1", "10"], fontsize=20)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=20)
plt.xlim([1e-4, 10])
plt.ylim([0, 1.2])
plt.tight_layout()
plt.savefig(fileloc + "coh_hr.svg", format="svg", dpi=1200)
plt.show()
```
